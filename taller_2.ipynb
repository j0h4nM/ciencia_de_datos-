```python

def plot_learning_curve(estimator,title,X,y,xlim=None,cv=None,n_jobs=1,train_sizes=np.linspace(.1, 1.0, 5)):
    plt.figure()
    plt.title(title)
    if ylim is not None:
        plt.ylim(*ylim)
    plt.xlabel("Ejemplos de entrenamiento")
    plt.ylabel("Puntuación de validación")
    train_sizes, train_scores, test_scores = learning_curve(
        estimator, X, y, cv=cv, n_jobs=n_jobs, train_sizes=train_sizes)
    train_scores_mean = np.mean(train_scores, axis=1)
    train_scores_std = np.std(train_scores, axis=1)
    test_scores_mean = np.mean(test_scores, axis=1)
    test_scores_std = np.std(test_scores, axis=1)
    plt.grid()

    plt.fill_between(train_sizes, train_scores_mean - train_scores_std,
                     train_scores_mean + train_scores_std, alpha=0.1,
                     color="r")
    plt.fill_between(train_sizes, test_scores_mean - test_scores_std,
                     test_scores_mean + test_scores_std, alpha=0.1, color="g")
    plt.plot(train_sizes, train_scores_mean, 'o-', color="r",
             label="Puntuación de entrenamiento")
    plt.plot(train_sizes, test_scores_mean, 'o-', color="g",
             label="Puntuación de validación")

    plt.legend(loc="best")
    return plt
```

```python
def plot_learning_curve(estimator, title, X, y, xlim=None, ylim=None, cv=None, n_jobs=1, train_sizes=np.linspace(.1, 1.0, 5)):


    # Creamos una nueva figura para la gráfica
    plt.figure()

    # Título del gráfico (texto que le das como parámetro)
    plt.title(title)

    # Si se especifican límites en el eje x, se aplican
    if xlim is not None:
        plt.xlim(*xlim)

    # Si se especifican límites en el eje y, se aplican
    if ylim is not None:
        plt.ylim(*ylim)

    # Etiquetas de los ejes
    plt.xlabel("Ejemplos de entrenamiento")
    plt.ylabel("Puntuación de validación")

    # Calculamos curvas de aprendizaje (con validación cruzada)
    # Devuelve el tamaño de los subconjuntos de entrenamiento y las puntuaciones correspondientes
    train_sizes, train_scores, test_scores = learning_curve(
        estimator, X, y, cv=cv, n_jobs=n_jobs, train_sizes=train_sizes)

    # Calculamos la media y desviación estándar de las puntuaciones
    train_scores_mean = np.mean(train_scores, axis=1)
    train_scores_std = np.std(train_scores, axis=1)
    test_scores_mean = np.mean(test_scores, axis=1)
    test_scores_std = np.std(test_scores, axis=1)

    # Activamos la cuadrícula del gráfico
    plt.grid()

    # Área sombreada para mostrar la desviación estándar de las puntuaciones de entrenamiento
    plt.fill_between(train_sizes,
                     train_scores_mean - train_scores_std,
                     train_scores_mean + train_scores_std,
                     alpha=0.1, color="r")

    # Área sombreada para la desviación estándar de las puntuaciones de validación
    plt.fill_between(train_sizes,
                     test_scores_mean - test_scores_std,
                     test_scores_mean + test_scores_std,
                     alpha=0.1, color="g")

    # Línea de la media de puntuaciones de entrenamiento
    plt.plot(train_sizes, train_scores_mean, 'o-', color="r",
             label="Puntuación de entrenamiento")

    # Línea de la media de puntuaciones de validación
    plt.plot(train_sizes, test_scores_mean, 'o-', color="g",
             label="Puntuación de validación")

    # Coloca la leyenda en la mejor posición automática
    plt.legend(loc="best")

    # Muestra la gráfica en pantalla (esto es esencial en Colab)
    plt.show()
```
