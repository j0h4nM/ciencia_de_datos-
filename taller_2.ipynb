###conclusiones
 1. Limpieza de Datos

Se eliminaron las filas con valores nulos en columnas clave (Age, Sex, Pclass, etc.), porque estas variables eran fundamentales para el análisis.

También se transformó la columna Age de float a int para facilitar las agrupaciones y visualizaciones por edad.

2. Agrupación por Edad
Dividimos la edad en grupos usando pd.cut() para entender mejor cómo influye la edad.

3. conclusiones

Sexo

Mujeres tenían una alta probabilidad de sobrevivir, mucho más que los hombres.

 Edad

Niños y adultos jóvenes (20-40 años) tenían mejores tasas de supervivencia.

Adultos mayores y adolescentes tenían tasas mucho más bajas.

Clase Social

Los pasajeros de primera clase sobrevivieron mucho más.

Los de tercera clase (estrato bajo) fueron los que más murieron.

Familiares a bordo

Tener 1 o 2 familiares (hermanos, esposos, padres o hijos) a bordo aumentaba la supervivencia.

Ir solo o con muchos bajaba las probabilidades.
